```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

# Курсовой проект

Для данных рейтингов по разным жанрам с сайта IMDB провести исследовательский анализ:
* Отобразить данные с помощью боксплота по каждому жанру, построить рядом для наглядности.
* Протестировать, влияет ли жанр на рейтинг. 
* С помощью линейной регрессии построить модель, и описать как рейтинг (величина у) зависит от различных переменных (хотя бы 2 взять).
* Предварительно провести корреляционный анализ и проверить данные на нормальность.

### Получим и подготавим данные

```python
data = pd.read_csv('imdb.csv', escapechar='\\')
gernes = ['Action', 'Adult', 'Adventure', 'Animation', 'Biography',
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy',
       'FilmNoir', 'GameShow', 'History', 'Horror', 'Music', 'Musical',
       'Mystery', 'News', 'RealityTV', 'Romance', 'SciFi', 'Short', 'Sport',
       'TalkShow', 'Thriller', 'War', 'Western']

data.dropna(inplace=True)
data.info()
```

### Отобразим данные с помощью боксплота по каждому жанру

```python
num_boxes = len(gernes)
rating = []

for i in gernes:
    rating.append(list(data[data[i]==1]['imdbRating']))

fig, ax1 = plt.subplots(figsize=(20, 20))
fig.canvas.set_window_title('')
fig.subplots_adjust(left=0.075, right=0.95, top=0.9, bottom=0.25)

bp = ax1.boxplot(rating)
plt.setp(bp['boxes'], color='black')
plt.setp(bp['whiskers'], color='gray')
plt.setp(bp['fliers'], color='red', marker='.')

ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',alpha=0.5)
ax1.set_axisbelow(True)
ax1.set_title('Анализ рейтингов по жанрам')
ax1.set_xlabel('Жанры')
ax1.set_ylabel('Рейтинг')

ax1.set_ylim(1, 10.5)
ax1.set_xlim(0.5, num_boxes + 0.5)
ax1.set_xticklabels(gernes, rotation=45, fontsize=10)

pos = np.arange(num_boxes) + 1
weights = ['bold', 'semibold']
box_colors = ['darkkhaki', 'royalblue']
for tick, label in zip(range(num_boxes), ax1.get_xticklabels()):
    k = tick % 2
    m = np.round(np.median(rating[pos[tick]-1]), 2)
    ax1.text(pos[tick], .97, m,
             transform=ax1.get_xaxis_transform(),
             horizontalalignment='center', 
             weight=weights[k], color=box_colors[k])


plt.show()
```

### Протестируем, влияет ли жанр на рейтинг.

```python
ym = []
y_all = []
for i in rating:
    ym.append(np.mean(i))
    y_all += i

k = len(rating)
y_mean = np.mean(y_all)
s2 = np.sum((y_all - y_mean)**2)

n = 0
s2_f = 0
s2_r = 0
for i in range(len(rating)):
    y = rating[i]
    n += len(y)
    s2_f += ((ym[i] - y_mean)**2) * len(y)
    s2_r += np.sum((y - ym[i])**2)

eta2 = s2_f / s2

f_val, p_val = stats.f_oneway(*rating)

print(f'Данные: Fh={f_val:0.5}, Pvalue = {p_val:0.5}, N2 = {eta2:0.5}')
```

> **<span style="color:red">Исходя из результата можно сделать вывод что жанр НЕ ВЛИЯЕТ на рейтинг</span>**

### Проведем корреляционный анализ и проверим данные на нормальность.

```python
y = list(data['imdbRating'])

stats.probplot(y, dist="norm", plot=plt)
plt.show()

stat, p = stats.shapiro(y)
print('Статистика=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p > alpha:
    print('Данные рейтинга являются нормальным распределением')
else:
    print('Данные рейтинга НЕ являются нормальным распределением')
```

```python
data[['imdbRating', 'ratingCount', 'duration', 'year']].corr()
```

> **Согласно результатам корреляции нет видимой линейной связи между рейтингом / количеством отзывов / длительностью и годом фильма**

### Построим модель линейной регрессии

```python
y = np.array(data[['imdbRating']])
n = len(y)
X = np.array(data[['ratingCount', 'duration']]) 
X = np.hstack([np.ones((n,1)) ,X])
B0, B1, B2 = np.dot(np.linalg.inv(np.dot(X.T, X)), X.T @ y).reshape((1,3))[0]

print(f'Уровнение линейной регрессии y = {B0:0.5} + {B1:0.5} * RatingCount + {B2:0.5} * Duration')
```

> **Очень слабая зависимость от количества отзывово и длительности фильма поэтому результаты такой модели могут далеки от правды**

```python
df = data[['imdbRating', 'ratingCount', 'duration']].copy()
df['Y_PRED'] = B0 + B1 * df['ratingCount'] + B2 * df['duration']
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(20, 8))
def draw(ax, field, title):
    ax.scatter(df[field], df['imdbRating'])
    ax.scatter(df[field], df['Y_PRED'], color="red")
    ax.set_xlabel(title)
    ax.set_ylabel('Рейтинг')

draw(axes[0], 'ratingCount', 'Количество обзоров')
draw(axes[1], 'duration', 'Длительность фильма')
plt.show()
```
