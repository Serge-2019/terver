```python
from math import factorial, sqrt
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

#### 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
salary = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

n = len(salary)
av = sum(salary)/n

smq = 0
for i in salary:
    smq += (i - av)**2
    
ds0 = smq / n
ds1 = smq / (n - 1)
    
st = sqrt(ds1)

print(f'среднее арифметическое = {av:0.5}')
print(f'среднее квадратичное отклонение = {st:0.5}')
print(f'смещенная оценка дисперсии = {ds0:0.5}')
print(f'несмещенная оценка дисперсии = {ds1:0.5}')
```

#### 2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
box1 = 8
box1w = 5
box1n = box1 - box1w # небелые
box2 = 12
box2w = 5
box2n = box2 - box2w # небелые
get1 = 2
get2 = 4

C1 = combinations(box1, get1) # все возможные сочетания для первого ящика
C2 = combinations(box2, get2) # все возможные сочетания для второго ящика

""" 
возможные вариации:

№ | первый яшик | второй яшик |
1 | 0 белых     | 3 белых    |
2 | 1 белых     | 2 белых    |
3 | 2 белых     | 1 белых    |
"""

def PP(b1w, b2w):
    p1 = combinations(box1w, b1w) * combinations(box1n, get1-b1w) / C1 # вариация для первого ящика
    p2 = combinations(box2w, b2w) * combinations(box2n, get2-b2w) / C2 # вариация для второго ящика
    return p1 * p2

P = PP(0,3) + PP(1,2) + PP(2,1)

print(f'Вероятность того, что 3 мяча белые = {P:0.5}')
```

#### 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: 
* a). первым спортсменом 
* б). вторым спортсменом 
* в). третьим спортсменом.

```python
def bayes(p1, p2, p3, pA1=1/3, pA2=1/3, pA3=1/3):
    psum = pA1 * p1 + pA2 * p2 + pA3 * p3
    p_1 = pA1 * p1 / psum
    p_2 = pA2 * p2 / psum
    p_3 = pA3 * p3 / psum
    return (p_1, p_2, p_3)

p_1, p_2, p_3 = bayes(0.9, 0.8, 0.6)

print(f'Выстрел произведен:\n\tпервым стрелком с вероятностью = {p_1:0.5}')
print(f'\tвторым стрелком с вероятностью = {p_2:0.5}')
print(f'\tтретьим стрелком с вероятностью = {p_3:0.5}')
```

#### 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: 
* a). на факультете A 
* б). на факультете B 
* в). на факультете C?

```python
p_A, p_B, p_C = bayes(0.8, 0.7, 0.9, 1/4, 1/4, 1/2)

print(f'Студент учиться:\n\tна факультете A = {p_A:0.5}')
print(f'\tна факультете B = {p_B:0.5}')
print(f'\tна факультете С = {p_C:0.5}')
```

#### 5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. 

* Какова вероятность того, что в первый месяц выйдут из строя: 
  * а). все детали 
  * б). только две детали 
  * в). хотя бы одна деталь 
  * г). от одной до двух деталей?

```python
A = 0.1 # выход из строя первой детали
Aw = 1-A # первая деталь работает нормально

B = 0.2 # выход из строя второй детали
Bw = 1-B # вторая деталь работает нормально

C = 0.25 # выход из строя третьей детали
Cw = 1-C # третья деталь работает нормально

all = A * B * C # все детали вышли из строя
onlytwo = A*B*Cw + C*B*Aw + A*C*Bw
onlyone = A*Bw*Cw + B*Aw*Cw + C*Aw*Bw

atleastone = onlyone + onlytwo + all
from1to2 = onlyone + onlytwo

print(f'Вероятность того, что в первый месяц выйдут из строя:')
print(f'\tвсе детали = {all:0.5}')
print(f'\tтолько две детали = {onlytwo:0.5}')
print(f'\tхотя бы одна деталь = {atleastone:0.5}')
print(f'\tот одной до двух деталей = {from1to2:0.5}')
```
