#### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

````python
a = 200
b = 800

M = (a + b) /2
D = (b - a)**2 / 12

print(f'Среднее значение = {M}')
print(f'Дисперсия = {D}')
````


Найдите:
* а). M(X)

````python
# так как функция плотности нормального распределения f(x) = 1/(o * sqrt(2*pi)) * exp(-(x-m) / 2*o^2 )
# то можно сделать вывод что 

# а) M(X) = -2 -  математическое ожидание
# б) D(X) = std(x)^2 = 4*4 = 16 - дисперсия
# в) std(X) = 4 - среднеквадратичное отклонение
````

#### 4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
* а). больше 182 см
* б). больше 190 см
* в). от 166 см до 190 см
* г). от 166 см до 182 см
* д). от 158 см до 190 см
* е). не выше 150 см или не ниже 190 см
* ё). не выше 150 см или не ниже 198 см
* ж). ниже 166 см.

````python
p182 = 0.1587 # больше 182 см Z=(182-174)/8 = 1 по таблице P = 0.1587 - это будет тоже самое что и для -1
print(f'а). больше 182 см \t\t\tP = {p182:0.5}')

p190 = 0.0228 # больше 190 см Z=(190-174)/8 = 2 по таблице P = 0.0228 - это будет тоже самое что и для -2
print(f'б). больше 190 см \t\t\tP = {p190:0.5}')

p166 = 0.1587 # ниже 166 см Z=(166-174)/8 = -1.0 по таблице P = 0.1587
res = (0.5 - p166) + (0.5 - p190) # складываем плотности от середины до 166 и от середины до 190
print(f'в). от 166 см до 190 см \t\t\tP = {res:0.5}')

res = (0.5 - p166) + (0.5 - p182) # складываем плотности от середины до 166 и от середины до 182
print(f'г). от 166 см до 182 см \t\t\tP = {res:0.5}')

p158 = 0.0228 # меньше 158 см Z=(158-174)/8 = -2 по таблице P = 0.0228
res = (0.5 - p158) + (0.5 - p190) # складываем плотности от середины до 158 и от середины до 190
print(f'д). от 158 см до 190 см \t\t\tP = {res:0.5}')

p150 = 0.0014 # меньше 150 см Z=(150-174)/8 = -3 по таблице P = 0.0014
res = p150 + p190 # складываем плотности ниже 158 и выше 190
print(f'е). не выше 150 см или не ниже 190 см \tP = {res:0.5}')

p198 = 0.0014 # выше 198 см Z=(198-174)/8 = 3 по таблице P = 0.0014 - - это будет тоже самое что и для -3
res = p150 + p198 # складываем плотности ниже 150 и выше 198
print(f'ё). не выше 150 см или не ниже 198 см \tP = {res:0.5}')

print(f'ж). ниже 166 см \t\t\tP = {p166:0.5}')
````

#### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

````python
M = 178
D = 25

o = sqrt(25)
print(f'Сигма равна o = {o:0.5}')
print(f'Рост человека отклоняется на {(190-M)/o} сигм')
````
